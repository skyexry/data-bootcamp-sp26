16. Define a function `pocket_change()` that takes four integers as inputs (numbers of pennies,
nickels, dimes, and quarters in your pocket) and returns a
floating point number (their dollar
value). Run your program with the input `(1, 2, 3, 4)`.

Bonus (optional): Report the
value with a dollar sign.

````python
#answer
def pocket_change(pennies, nickels, dimes, quarters):
  total = (pennies * 0.01 +
           nickels * 0.05 +
           dimes * 0.10 +
           quarters * 0.25)
  return total

value = pocket_change(1, 2, 3, 4)
print(f"${value:.2f}")
````

17. Write a function `notsix()` that takes a list of integers and returns a (shorter)
list of only those that do not begin with a 6. Test it on the list `[1234, 6783, 6, 4321,
9876]`.

````python
#answer
def notsix(inlist = []):
  result = []
  for n in inlist:
    s = str(n)
    if s[0] != '6':
      result.append(n)
  return result
````

````python
#test
assert notsix([1234, 6783, 6, 4321, 9876]) == [1234, 4321, 9876]
````
