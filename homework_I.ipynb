**Problem 19**:  Code Wars Problem I

Write a function that accepts an array of 10 integers (each between 0 and 9) and returns a string formatted as a phone number.

Example

`createPhoneNumber([1, 2, 3, 4, 5, 6, 7, 8, 9, 0])`  
returns: `"(123) 456-7890"`

Requirements

- The returned format must be exactly correct.
- Don't forget the space after the closing parentheses: `") "`


````python
def createPhoneNumber(numbers):
    return f"({numbers[0]}{numbers[1]}{numbers[2]}) {numbers[3]}{numbers[4]}{numbers[5]}-{numbers[6]}{numbers[7]}{numbers[8]}{numbers[9]}"

print(createPhoneNumber([1,2,3,4,5,6,7,8,9,0]))  # "(123) 456-7890"
````

**Problem 20**: Code Wars Problem II

The maximum sum subarray problem consists in finding the maximum sum of a contiguous subsequence in an array or list of integers.

For example:

Input: [-2, 1, -3, 4, -1, 2, 1, -5, 4]  
Output: 6 (Sum of [4, -1, 2, 1])

Easy case is when the list is made up of only positive numbers and the maximum sum is the sum of the whole array. If the list is made up of only negative numbers, return 0 instead. Your solution should be fast, it will be tested on very large arrays so slow solutions will time out.

Empty list is considered to have zero greatest sum. Note that the empty list or array is also a valid sublist/subarray.


````python
def max_sequence(arr):
    best = 0
    curr = 0
    for x in arr:
        curr = max(0, curr + x)
        best = max(best, curr)
    return best

print(max_sequence([-2, 1, -3, 4, -1, 2, 1, -5, 4]))  # 6
````
